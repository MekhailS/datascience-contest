```python
import pandas as pd
DATA_PROC_PATH = 'data/vehicle_proccessed/'
vehicle1_summary = pd.read_csv(DATA_PROC_PATH + 'vehicle1_summary.csv', sep=',')
```

```python
def replaceNanValue(arr):
    new = []
    for i in arr:
        if not np.isnan(i):
            new.append(i)
        else:
            new.append(0)
    for i in new[: : -1]:
        if i == 0:
            new.pop()
        else:
            break
    return new
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(50,3))
y = replaceNanValue(vehicle1_summary['fuellevel'])
x = np.arange(len(y))
plt.plot(x, y)
plt.show
```
